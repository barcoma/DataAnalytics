```python
dates = greentrip[:,0:2]
ratecode = greentrip[:,2].astype(np.int16)
trip_from = greentrip[:,3]
trip_to = greentrip[:,4]
passengers = greentrip[:,5].astype(np.int16)
distance = greentrip[:,6].astype(np.float64)
cost = greentrip[:,7].astype(np.float64)
```

```python
costpermile = cost.mean() / distance.mean()
print("Der durchschnittliche Preis/Meile beträgt: ")
print(costpermile)
```

```python
max_distance = np.argmax(distance)
trip_to[max_distance]
trip_from [max_distance]
print("Die längste Strecke wurde von "+trip_from[max_distance]+" nach "+trip_to[max_distance]+"zurückgelegt und betrug: ")
print(distance[max_distance])
```

```python
filter_gtOne = passengers > 1
cost_gtOne = cost[filter_gtOne]
distance_gtOne = distance[filter_gtOne]
costpermile_gtOne = cost_gtOne.mean() / distance_gtOne.mean()
print("Der durchschnittliche Preis/Meile beträgt: ")
print(costpermile_gtOne)
```
